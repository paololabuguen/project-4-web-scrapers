```python
# List of columns to remove for loan default prediction
columns_to_remove = ['ID', 'Funded Amount', 'Funded Amount Investor', 'Term',
                     'Batch Enrolled', 'Grade', 'Sub Grade', 'Payment Plan',
                     'Loan Title', 'Public Record', 'Collection 12 months Medical',
                     'Application Type', 'Last week Pay', 'Accounts Delinquent']

df_clean = Data.drop(columns=columns_to_remove)
```

5.Handling catagorical columns

```python
df_clean.sample(5)
```

```python
# Data encoding
df_clean['Home Ownership'] = df_clean['Home Ownership'].map({'MORTGAGE': 1, 'RENT': 2, 'OWN': 3}).astype('int')
df_clean["Verification Status"] = df_clean["Verification Status"].map({"Not Verified": 1, "Source Verified": 2, "Verified": 3}).astype('int')
df_clean["Initial List Status"] = df_clean["Initial List Status"].map({"w": 1, "f": 2}).astype('int')

print("Current shape of dataset:", df_clean.shape)
```

```python
df_clean.sample(5)
```

7.Store feauture matrix in X and response in y

```python
X = df_clean.drop('Loan Status', axis = 1)
y = df_clean['Loan Status']
```

8.Feature scaling

```python
df_clean.head()
```

```python
cols = ['Loan Amount', 'Interest Rate', 'Employment Duration (years)', 'Debit to Income', 
        'Open Account', 'Revolving Balance', 'Total Accounts', 'Total Received Interest', 'Total Received Late Fee', 
        'Recoveries', 'Collection Recovery Fee', 'Total Collection Amount', 'Total Current Balance', 'Total Revolving Credit Limit']
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X[cols]=st.fit_transform(X[cols])
```

```python
X
```

9.Splitting the data into test and training model 

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
model_df = {}
def model_val(model,X,y):
    X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                     test_size = 0.20,
                                                     random_state =42)
    
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print(f"{model} acuracy is {accuracy_score(y_test,y_pred)}")

    score = cross_val_score(model,X,y,cv = 5)
    print(f"{model} Avg corss val score is {np.mean(score)}")
    model_df[model] = round(np.mean(score)*100,2)
```

10.Logistic regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model_val(model,X,y)
```

12.Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model_val(model,X,y)
```

13.Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model_val(model,X,y)
```

14.Gradient boosting classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model_val(model,X,y)
```

15.Hyperparameter tunning

```python
from sklearn.model_selection import RandomizedSearchCV
```

Logistic regression

```python
log_reg_grid = {"C":np.logspace(-4,4,20),
                "solver":['liblinear']}
```

```python
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                   param_distributions = log_reg_grid,
                   n_iter = 20, cv = 5, verbose = True)
```

```python
rs_log_reg.fit(X,y)
```
